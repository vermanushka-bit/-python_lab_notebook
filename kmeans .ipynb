```python


# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


# Create Random Dataset

np.random.seed(42)

study_hours = np.random.randint(1, 10, 20)
marks = np.random.randint(30, 100, 20)

# Create DataFrame
df = pd.DataFrame({
    "Study_Hours": study_hours,
    "Marks": marks
})

# Display dataset
print("RANDOM DATASET\n")
print(df)


# Select Features


X = df[["Study_Hours", "Marks"]]


# Apply K-Means Clustering


# Create model with 3 clusters
model = KMeans(n_clusters=3, random_state=42)

# Train model
model.fit(X)

# Predict clusters
df["Cluster"] = model.predict(X)

# Display clustered dataset
print("\nDATASET WITH CLUSTERS\n")
print(df)


# Display Centroids

print("\nCentroids:")
print(model.cluster_centers_)


# Scatter Plot

plt.scatter(df["Study_Hours"],
            df["Marks"],
            c=df["Cluster"])

# Plot centroids
plt.scatter(model.cluster_centers_[:, 0],
            model.cluster_centers_[:, 1],
            marker='X',
            s=200)

plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("K-Means Clustering using Random Dataset")

plt.grid(True)
plt.show()
```
