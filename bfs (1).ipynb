```python
from collections import deque
def bfs(graph, start, goal):
    visited = []
    queue = deque([[start]])   # store path
    
    while queue:
        path = queue.popleft()
        node = path[-1]
        
        if node not in visited:
            visited.append(node)
            
            # Goal found
            if node == goal:
                return visited, path
            
            # Explore neighbors
            for neighbour in graph[node]:
                new_path = list(path)
                new_path.append(neighbour)
                queue.append(new_path)
    
    return visited, []

```

```python
from collections import deque
print("\n===== DYNAMIC GRAPH =====")

n = int(input("Enter number of nodes: "))
nodes = []

for i in range(n):
    node = input(f"Enter node {i+1}: ").upper()
    nodes.append(node)

graph_dynamic = {node: [] for node in nodes}

e = int(input("Enter number of edges: "))
print("Enter edges (node1 node2):")

for _ in range(e):
    u, v = input().upper().split()

    if u in graph_dynamic and v in graph_dynamic:
        graph_dynamic[u].append(v)
        graph_dynamic[v].append(u)   # undirected
    else:
        print("Invalid edge skipped!")

# Input start and goal
while True:
    start = input("Enter start node: ").upper()
    if start in graph_dynamic:
        break
    print("Invalid node!")

while True:
    goal = input("Enter goal node: ").upper()
    if goal in graph_dynamic:
        break
    print("Invalid node!")

# BFS Traversal
visited, path = bfs(graph_dynamic, start, goal)

print("Traversal Order:", visited)
print("Shortest Path:", path if path else "No path found")
```
