```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create dataset using dictionary
data = {
    "Student_Name": ["Aman", "Riya", "Kunal", "Sneha", "Rahul",
                     "Priya", "Aditi", "Vikas", "Neha", "Arjun"],

    "Sem": [1, 2, 1, 3, 1, 2, 4, 1, np.nan, 2],

    "Age": [18, 19, 18, 20, 19, np.nan, 21, 18, 20, 19],

    "Height": [5.6, 5.4, 5.8, np.nan, 5.7, 5.5, 5.3, 6.5, 5.4, 5.8],

    "Study_Hours": [2, 4, 3, 6, np.nan, 5, 7, 2, 4, 8],

    "Marks": [45, 65, 55, 78, 49, 72, 88, 150, np.nan, 95]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display dataset
print("DATASET\n")
print(df)


# i. Value count of students in first year (Sem = 1)

first_year_count = (df["Sem"] == 1).sum()

print("\nNumber of students in First Year:")
print(first_year_count)


# ii. Outliers in Marks column using IQR method


Q1 = df["Marks"].quantile(0.25)
Q3 = df["Marks"].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

outliers = df[(df["Marks"] < lower_limit) |
              (df["Marks"] > upper_limit)]

print("\nOutliers in Marks Column:")
print(outliers[["Student_Name", "Marks"]])


# iii. Correlation between Marks and Study Hours

correlation = df["Marks"].corr(df["Study_Hours"])

print("\nCorrelation between Study Hours and Marks:")
print(correlation)


# iv. Line Graph between Study Hours and Marks

# Remove null values for graph
graph_data = df.dropna(subset=["Study_Hours", "Marks"])

plt.plot(graph_data["Study_Hours"],
         graph_data["Marks"],
         marker='o')

plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")

plt.grid(True)
plt.show()
```
